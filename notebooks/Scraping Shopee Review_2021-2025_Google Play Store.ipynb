```python
import pandas as pd
import time
from tqdm import tqdm
from google_play_scraper import reviews, Sort

APP_ID = "com.shopee.id"

START_DATE = pd.to_datetime("2025-01-01")
END_DATE    = pd.to_datetime("2025-12-31")

BATCH_SIZE = 200
DELAY = 1.5

all_reviews = []
continuation_token = None

print("Scraping based on timeline...")

last_displayed_month = None # To track the month displayed in the progress bar description

with tqdm(desc="Scraping reviews", unit=" reviews") as pbar: # Initialize tqdm with a default description
    while True:

        result, continuation_token = reviews(
            APP_ID,
            lang='id',
            country='id',
            sort=Sort.NEWEST,
            count=BATCH_SIZE,
            continuation_token=continuation_token
        )

        if not result:
            break

        stop_scraping = False
        batch_reviews_collected = [] # To store reviews relevant to the date range and content filter

        for r in result:
            review_date = pd.to_datetime(r["at"])

            # ON RANGE > SAVE AND CHECK FOR CONTENT
            if START_DATE <= review_date <= END_DATE and r["content"]:
                batch_reviews_collected.append({
                    "review": r["content"],
                    "rating": r["score"],
                    "date": review_date,
                    "thumbs_up": r["thumbsUpCount"]
                })

            # OUT OF RANGE → STOP
            elif review_date < START_DATE:
                stop_scraping = True
                break

        all_reviews.extend(batch_reviews_collected) # Add collected reviews from this batch to the main list

        # Update the progress bar description with the current month being processed
        # This will show the month of the oldest review collected in the current batch
        if batch_reviews_collected:
            current_oldest_review_date = batch_reviews_collected[-1]['date']
            current_month_str = current_oldest_review_date.strftime('%Y-%m')

            if current_month_str != last_displayed_month:
                pbar.set_description(f"Scraping reviews (currently in {current_month_str})")
                last_displayed_month = current_month_str

        pbar.update(len(result)) # Update the progress bar with the total count of reviews fetched by the scraper

        if stop_scraping:
            print("Reached start date limit. Stopping...")
            break

        time.sleep(DELAY)

        if continuation_token is None:
            break

df_raw = pd.DataFrame(all_reviews)

print("Total data (Jan–Dec 2025):", len(df_raw))
df_raw.head()
```

### Melanjutkan Proses Scraping

Untuk melanjutkan scraping, Anda perlu melakukan dua hal:
1.  **Muat ulang data parsial** yang sudah disimpan.
2.  **Muat ulang `continuation_token`** terakhir (jika sudah disimpan). Jika tidak, proses akan dimulai dari awal namun akan menghindari duplikasi dengan memfilter berdasarkan tanggal terbaru dari data yang sudah ada.

```python
import pandas as pd
import time
from tqdm import tqdm
from google_play_scraper import reviews, Sort
import re # Import re for regex

# APP_ID, BATCH_SIZE, DELAY should be defined from previous cells
# START_DATE and END_DATE for 2022 were defined in cell GdbYFgGC_bf7

# --- 1. Muat ulang data parsial ---
# Asumsi Anda telah menyimpan file '2022_shopee_reviews_raw_partial.csv' ke Google Drive
# Jika Anda tidak menyimpan ke Drive, ganti path sesuai lokasi lokal Anda

# Mount Google Drive jika belum ter-mount
from google.colab import drive
drive.mount('/content/gdrive')

drive_path_partial = '/content/gdrive/My Drive/2022_shopee_reviews_raw_partial.csv'

try:
    # Coba muat data parsial dari Drive
    all_reviews = pd.read_csv(drive_path_partial).to_dict('records')
    print(f"Data parsial berhasil dimuat: {len(all_reviews)} ulasan.")
    # Sort by date in descending order to find the latest review date
    latest_review_date_in_partial = pd.to_datetime(all_reviews[0]['date']) if all_reviews else START_DATE # Use START_DATE if no reviews
    print(f"Ulasan terakhir dalam data parsial: {latest_review_date_in_partial}")

except FileNotFoundError:
    print("File data parsial tidak ditemukan di Google Drive. Memulai ulang scraping.")
    all_reviews = [] # Mulai dari kosong jika file tidak ditemukan
except Exception as e:
    print(f"Gagal memuat data parsial: {e}. Memulai ulang scraping.")
    all_reviews = []


# --- 2. Muat ulang continuation_token (jika disimpan) ---
# Path ke file continuation_token di Google Drive
drive_path_token = '/content/gdrive/My Drive/2022_continuation_token.txt'
continuation_token_to_resume = None

try:
    with open(drive_path_token, 'r') as f:
        token_str = f.read().strip()
        if token_str and token_str != 'None':
            # CHECK if the token is saved as the object's repr string
            if re.match(r"<google_play_scraper\.features\.reviews\._ContinuationToken object at 0x[0-9a-fA-F]+>", token_str):
                print(f"WARNING: The saved continuation token is in an incorrect format (object representation). Sayangnya, saya tidak bisa melanjutkan proses scraping menggunakan format token yang tidak valid ini.")
                print(f"SOLUSI: Mohon jalankan kembali sel 187ba802 dan a7c08411 setelah memodifikasi sel 187ba802 untuk menyimpan 'continuation_token.token' (bukan 'str(continuation_token)').")
                print("Scraping tidak dapat dilanjutkan dengan format token ini.")
                continuation_token_to_resume = None # Set to None to prevent further error
            else:
                # If it's not the repr string, assume it's the actual token string
                # The google_play_scraper.reviews function can accept the raw token string
                continuation_token_to_resume = token_str
                print(f"Continuation token dimuat: {continuation_token_to_resume[:30]}...")
        else:
            print("Tidak ada continuation token yang tersimpan atau token adalah 'None'.")
except FileNotFoundError:
    print("File continuation token tidak ditemukan. Melanjutkan tanpa token.")
except Exception as e:
    print(f"Gagal memuat continuation token: {e}. Melanjutkan tanpa token.")

# Set continuation_token untuk loop baru
continuation_token = continuation_token_to_resume

# Untuk menghindari duplikasi, set START_DATE baru untuk scraping
# Kita akan mencari ulasan yang lebih baru dari ulasan terbaru yang sudah ada
# Jika tidak ada continuation_token yang dimuat, maka START_DATE harus diatur secara hati-hati
# agar tidak mengulang data yang sudah ada di 'all_reviews'
if all_reviews and not continuation_token_to_resume:
    # Jika hanya data yang dimuat, dan tidak ada token, sesuaikan START_DATE untuk menghindari duplikasi
    # Kita akan set START_DATE menjadi sedikit setelah ulasan terbaru yang kita miliki
    # Namun, karena scraper mengambil berdasarkan `Sort.NEWEST`, dengan token jauh lebih efektif.
    # Tanpa token, kita akan tetap mendapatkan ulasan baru dan filter secara manual.
    print("Memuat data tanpa continuation_token. Filtering berdasarkan tanggal akan dilakukan.")


# --- 3. Lanjutkan scraping ---
print("Melanjutkan scraping...")

last_displayed_month = None

# Only attempt to scrape if a valid continuation_token is available
if continuation_token is not None:
    with tqdm(desc="Scraping reviews (lanjutan)", unit=" reviews") as pbar:
        while True:
            try:
                # If continuation_token_to_resume was a string token, pass it directly
                # If it was a _ContinuationToken object (which it won't be if loaded from file this way),
                # google_play_scraper.reviews can handle it.
                # However, the problem was that `continuation_token` was the repr string.
                # By setting continuation_token_to_resume = None if the format is bad,
                # we prevent the error, but also stop scraping.
                result, new_continuation_token_object = reviews(
                    APP_ID,
                    lang='id',
                    country='id',
                    sort=Sort.NEWEST,
                    count=BATCH_SIZE,
                    continuation_token=continuation_token # This should now be either a string token or None
                )
                continuation_token = new_continuation_token_object # Update to the actual object for next iteration

            except Exception as e:
                print(f"Error saat melanjutkan scraping: {e}. Menghentikan.")
                break

            if not result:
                print("Tidak ada ulasan baru. Menghentikan.")
                break

            stop_scraping = False
            batch_reviews_collected = []

            for r in result:
                review_date = pd.to_datetime(r["at"])

                # ON RANGE > SAVE AND CHECK FOR CONTENT
                # Hanya tambahkan ulasan jika dalam rentang tanggal dan ada konten, DAN belum ada di all_reviews
                # Filter berdasarkan review_date > latest_review_date_in_partial jika tidak ada continuation_token_to_resume
                # Namun, continuation_token sudah otomatis mengatasi ini.
                if START_DATE <= review_date <= END_DATE and r["content"]:
                    # Cek duplikasi jika continuation_token tidak dimuat atau ada kekhawatiran duplikasi
                    # Dengan continuation_token, ini seharusnya tidak terjadi
                    # Jika tidak menggunakan token, Anda mungkin perlu mengecek apakah ulasan ini sudah ada di all_reviews
                    new_review = {
                        "review": r["content"],
                        "rating": r["score"],
                        "date": review_date,
                        "thumbs_up": r["thumbsUpCount"]
                    }
                    batch_reviews_collected.append(new_review)

                # OUT OF RANGE → STOP
                elif review_date < START_DATE:
                    stop_scraping = True
                    break

            all_reviews.extend(batch_reviews_collected)

            if batch_reviews_collected:
                current_oldest_review_date = batch_reviews_collected[-1]['date']
                current_month_str = current_oldest_review_date.strftime('%Y-%m')

                if current_month_str != last_displayed_month:
                    pbar.set_description(f"Scraping reviews (lanjutan, currently in {current_month_str})")
                    last_displayed_month = current_month_str

            pbar.update(len(result))

            if stop_scraping:
                print("Mencapai batas tanggal awal yang ditentukan. Menghentikan.")
                break

            time.sleep(DELAY)

            if continuation_token is None:
                print("Continuation token menjadi None. Menghentikan.")
                break
else:
    print("Scraping tidak dapat dilanjutkan karena continuation token tidak valid atau tidak ditemukan.")


df_raw = pd.DataFrame(all_reviews)

print(f"Total data (termasuk yang dimuat dan baru disekrap): {len(df_raw)}")
display(df_raw.head())
```

```python
# Data handling
import pandas as pd
import numpy as np

# Scraping
from google_play_scraper import reviews, Sort

# Utility
import time
from tqdm import tqdm
import re
```

```python
APP_ID = "com.shopee.id"
TARGET_REVIEWS = 1_000_000
BATCH_SIZE = 200
DELAY = 1.5

print("Config ready!")
```

```python
def scrape_shopee_reviews(app_id, target_reviews=1000000):
    all_reviews = []
    continuation_token = None

    with tqdm(total=target_reviews) as pbar:
        while len(all_reviews) < target_reviews:
            try:
                result, continuation_token = reviews(
                    app_id,
                    lang='id',
                    country='id',
                    sort=Sort.NEWEST,
                    count=BATCH_SIZE,
                    continuation_token=continuation_token
                )
            except Exception as e:
                print("Error saat scraping:", e)
                break

            if not result:
                break

            for r in result:
                all_reviews.append({
                    "review": r["content"],
                    "rating": r["score"],
                    "date": r["at"],
                    "thumbs_up": r["thumbsUpCount"]
                })

            pbar.update(len(result))
            time.sleep(DELAY)

            if continuation_token is None:
                break

    return pd.DataFrame(all_reviews)
```
