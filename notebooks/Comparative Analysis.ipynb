```python
import pandas as pd
import matplotlib.pyplot as plt

comparison = pd.DataFrame({
    "Model": [
        "TD-LSTM",
        "BiLSTM",
        "GRU",
        "Attention\nBiLSTM",
        "IndoBERT"
    ],
    "Accuracy": [
        74.78,
        89.59,
        88.83,
        89.71,
        91.34
    ]
})

plt.figure(figsize=(8,5))

bars = plt.bar(
    comparison["Model"],
    comparison["Accuracy"]
)

plt.ylabel("Accuracy (%)")
plt.title("Comparison of Accuracy Across Models")
plt.ylim(70,95)

for bar in bars:
    y = bar.get_height()
    plt.text(
        bar.get_x()+bar.get_width()/2,
        y+0.2,
        f"{y:.2f}",
        ha="center"
    )

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ===========================
# DATA HASIL PENGUJIAN
# ===========================

comparison = pd.DataFrame({
    "Model": [
        "TD-LSTM",
        "GRU",
        "BiLSTM",
        "Attention BiLSTM",
        "IndoBERT"
    ],

    "Accuracy": [
        74.78,
        88.83,
        89.59,
        89.71,
        91.34
    ],

    "Precision": [
        81.02,
        85.53,
        86.25,
        86.40,
        91.67
    ],

    "Recall": [
        74.78,
        88.83,
        89.59,
        89.71,
        91.34
    ],

    "F1-score": [
        76.76,
        87.09,
        87.88,
        87.95,
        90.66
    ]
})

# ===========================
# LINE CHART
# ===========================

plt.figure(figsize=(10,6))

plt.plot(
    comparison["Model"],
    comparison["Accuracy"],
    marker='o',
    linewidth=2,
    label="Accuracy"
)

plt.plot(
    comparison["Model"],
    comparison["Precision"],
    marker='s',
    linewidth=2,
    label="Precision"
)

plt.plot(
    comparison["Model"],
    comparison["Recall"],
    marker='^',
    linewidth=2,
    label="Recall"
)

plt.plot(
    comparison["Model"],
    comparison["F1-score"],
    marker='D',
    linewidth=2,
    label="F1-score"
)

# ===========================

for metric in ["Accuracy","Precision","Recall","F1-score"]:
    for x, y in zip(comparison["Model"], comparison[metric]):
        plt.text(
            x,
            y+0.3,
            f"{y:.2f}",
            fontsize=8,
            ha='center'
        )

plt.title("Performance Comparison of Sentiment Classification Models")

plt.xlabel("Model")

plt.ylabel("Score (%)")

plt.ylim(70,95)

plt.grid(True, linestyle="--", alpha=0.5)

plt.legend()

plt.tight_layout()

plt.savefig(
    "comparison_models.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
